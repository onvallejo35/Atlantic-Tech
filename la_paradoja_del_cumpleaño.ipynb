# 🎂 La Paradoja del Cumpleaños

Simulación para estimar la probabilidad de que al menos dos personas en un grupo de k personas cumplan años el mismo día.

**Autor:** Nadiu.
**Método:** Simulación con N = 1000 repeticiones


## 🧠 Idea

Se generan fechas de cumpleaños aleatorias (entre 1 y 365) para k personas.
Luego se verifica si hay coincidencias.
Esto se repite N veces para estimar la probabilidad.


```python
import random
import matplotlib.pyplot as plt
```

## 🔹 1. Experimento aleatorio

```python
def cumples(k):
    return [random.randint(1, 365) for _ in range(k)]
```

## 🔹 2 y 3. Función filtro

```python
def hay_coincidencia(lista_cumples):
    return len(lista_cumples) != len(set(lista_cumples))
```

## 🔹 4. Aproximación de la probabilidad

```python
def proporcion_coincidencia(k, N=1000):
    random.seed(42)
    coincidencias = 0

    for _ in range(N):
        muestras = cumples(k)
        if hay_coincidencia(muestras):
            coincidencias += 1

    return coincidencias / N
```

## 📊 Resultados para k = 1 a 70

```python
resultados = []

for k in range(1, 71):
    prob = proporcion_coincidencia(k, 1000)
    resultados.append(prob)
    print(f"k = {k}, Probabilidad ≈ {prob:.4f} ({prob*100:.2f}%)")
```

## 📈 Gráfico

```python
k_values = list(range(1, 71))

plt.figure(figsize=(14, 8))
plt.plot(k_values, resultados,marker='o', linestyle='-', label='Simulación (Monte Carlo)')
plt.xlabel("Cantidad de personas (k)")
plt.ylabel("Probabilidad de coincidencia")
plt.title("Paradoja del Cumpleaños")
plt.axhline(y=0.5, color='r', linestyle='--', label='Probabilidad 0.5')
plt.axvline(x=23, color='g', linestyle=':', label='k=23 (Punto crítico)')
plt.legend()
plt.grid()
plt.show()
```

## 🧾 Conclusión

La simulación muestra que la probabilidad de coincidencia crece rápidamente.
Alrededor de k = 23 personas, la probabilidad ya supera el 50% y alrededor de k = 70 personas la probabilidad es del 100%

Esto ocurre porque se evalúan todas las combinaciones posibles de pares dentro del grupo.
